```python
# 使用 Docker 快速搭建 PySpark 环境
# https://medium.com/@bee811101/%E4%BD%BF%E7%94%A8-docker-%E5%BF%AB%E9%80%9F%E5%BB%BA%E7%BD%AE-pyspark-%E7%92%B0%E5%A2%83-657f9d8bff3a

from pyspark import SparkContext

sc = SparkContext("local", "First App")

# 載入檔案
text_file = sc.textFile("fs-auth.md")

# word count
counts = text_file.flatMap( lambda line: line.strip().lower().split(" ") ) \
            .filter( lambda word : word ) \
            .map( lambda word: (word, 1) ) \
            .reduceByKey( lambda a, b: a + b ) \
            .sortBy( lambda x: x[1], False )
output = counts.collect()

# 印出字與數量
for (word, count) in output:
    print( "%s: %i" % (word, count) )
    
# Stopping Spark Context
sc.stop()
```
