```python
from diagrams import Cluster, Diagram, Node, Edge

no_padding = {
    "pad": "0",
    "nodesep": "0.12",
    # "ranksep": "0",
}
blue, pink, green, purple, yellow = "#ADD8E6", "#FFD9D9", "#D8E2D8", "#E7D9F5", "#F0E8D0"
lblue, lgreen, lpurple, lyellow, lpink = "#E5F5FD", "#EBF3E7", "#ECE8F6", "#FDF7E3", "#FFEBEB"
node_attr = { "labelloc": "c", "height": "0.4", "style": "rounded,filled", "fillcolor": lyellow }

class Align(Edge):
    def __init__(self):
        super().__init__(penwidth="0")
align = Align()

class Item(Node):
    def __init__(self, label, **kwargs):
        super().__init__(label, **{ **node_attr, **kwargs })
class Blank(Item):
    def __init__(self):
        item_attr = { "penwidth": "0", "fillcolor": "transparent" }
        super().__init__("", **{ **node_attr, **item_attr })

with Diagram("", show=False, graph_attr=no_padding) as diag:
    with Cluster("个人规划", graph_attr={"bgcolor":lblue}):

        with Cluster("收入", graph_attr={"bgcolor":pink}):
            a0 = Item("简历求职", fillcolor=lpink)
            Item("技术咨询", fillcolor=lpink)
            Item("创业产品", fillcolor=lpink)
            a1 =Blank()
    
        with Cluster("能力", graph_attr={"bgcolor":green}):
            b0 = Item("高并发/高可用/云原生", fillcolor=lgreen, width="2")
            Item("微服务/中间件/容器", fillcolor=lgreen, width="2")
            Item("信息系统/主机运维", fillcolor=lgreen, width="2")
            b1 = Item("计算机基础/编程语言", fillcolor=lgreen, width="2")

        with Cluster("背书", graph_attr={"bgcolor":purple}):
            c0 = Item("开源项目", fillcolor=lpurple)
            Item("产品集", fillcolor=lpurple)
            Item("作品集", fillcolor=lpurple)
            c1 = Blank()

    a0 - align - b0 - align - c0
    a1 - align - b1 - align - c1
diag
```
