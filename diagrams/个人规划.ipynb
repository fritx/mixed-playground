```python
from diagrams import Cluster, Diagram, Node, Edge

no_padding = {
    "pad": "0",
    "nodesep": "0.12",
    # "ranksep": "0",
}
blue, pink, green, purple, yellow = "#ADD8E6", "#FFD9D9", "#D8E2D8", "#E7D9F5", "#F0E8D0"
lblue, lgreen, lpurple, lyellow, lpink = "#E5F5FD", "#EBF3E7", "#ECE8F6", "#FDF7E3", "#FFEBEB"
node_attr = { "labelloc": "c", "height": "0.4", "style": "rounded,filled", "fillcolor": lyellow }

class Align(Edge):
    def __init__(self):
        super().__init__(penwidth="0")
align = Align()

class Item(Node):
    def __init__(self, label, **kwargs):
        super().__init__(label, **{ **node_attr, **kwargs })
class Blank(Item):
    def __init__(self):
        item_attr = { "penwidth": "0", "fillcolor": "transparent" }
        super().__init__("", **{ **node_attr, **item_attr })

with Diagram("", show=False, graph_attr=no_padding) as diag:
    with Cluster("", graph_attr={"bgcolor":lblue}):

        with Cluster("", graph_attr={"bgcolor":pink}):
            a0 = Item("", fillcolor=lpink)
            Item("", fillcolor=lpink)
            Item("", fillcolor=lpink)
            a1 =Blank()
    
        with Cluster("", graph_attr={"bgcolor":green}):
            b0 = Item("//", fillcolor=lgreen, width="2")
            Item("//", fillcolor=lgreen, width="2")
            Item("/", fillcolor=lgreen, width="2")
            b1 = Item("/", fillcolor=lgreen, width="2")

        with Cluster("", graph_attr={"bgcolor":purple}):
            c0 = Item("", fillcolor=lpurple)
            Item("", fillcolor=lpurple)
            Item("", fillcolor=lpurple)
            c1 = Blank()

    a0 - align - b0 - align - c0
    a1 - align - b1 - align - c1
diag
```
